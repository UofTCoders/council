```python
import pandas as pd
```

```python
track_not_nan = pd.read_csv('./tracks-clean-no-nans.csv', index_col=0)
print(track_not_nan.info())
track_not_nan.head()
```

```python
echo = pd.read_csv('./echonest-clean.csv', index_col=0)
print(echo.info())
echo.head()
```

```python
track_not_nan['genre_top'].value_counts()
```

```python
echo_tracks = echo.join(track_not_nan['genre_top'])
echo_tracks = echo_tracks.loc[echo_tracks['genre_top'].isin(['Hip-Hop', 'Rock'])]
echo_tracks.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

tree = DecisionTreeClassifier()

train_data, test_data, train_label, test_label = train_test_split(echo_tracks.iloc[:, :-1], echo_tracks['genre_top'])
tree.fit(train_data, train_label)
tree.score(test_data, test_label)
```
